## Fixando a_0 positivo, variando a_1

```python
X_init = np.array([1,1])
print(f"X_init = {X_init}\n")
a_0  = -3.4
plt.figure(dpi=120)
for idx_a,a_1 in enumerate(sorted(np.linspace(10,-10,9))):
    A = get_A([a_0,a_1])
    system = DynamicSystem(A,X_init)

    det_a_0 = system.det
    test_a_0 = det_a_0<0
    t, X_0, X_1 = system.get_stability(t=np.linspace(0,10,10000), verbose=False)
    X_dot = system.X_dot

    color=f"C{idx_a}"
    plt.plot(X_0, X_1,label='$a_1$ ='+f' {a_1:.3f}',color=color)
    
    plt.arrow(X_init[0], X_init[1] , X_dot[0]*0.03, X_dot[1]*0.03,lw=0.45, width = .01,color=color, zorder=10,ec='black')
    
plt.title("Sela ($a_0$ = "+f"{a_0:.2f}"+" e $a_1$ variando)")
plt.grid()
plt.legend()
plt.xlim([X_init[0]-0.5,X_init[0]+1])
plt.ylim([X_init[1]-1,4])
plt.xlabel("$X_0$")
plt.ylabel("$X_1$")
plt.savefig("img/a_0_var_a_1.png")
```
